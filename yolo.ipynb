```python
 import cv2
```

```python
def crop_image_yolo(image_path, coordinates):
    image = cv2.imread(image_path)

    # Extract coordinates
    x1, y1, x2, y2 = map(int, coordinates)

    # Crop the image
    cropped_image = image[y1:y2, x1:x2]

    return cropped_image
```
